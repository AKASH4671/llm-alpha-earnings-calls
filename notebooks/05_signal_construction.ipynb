```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
data = pd.read_pickle("../data/processed/transcript_embeddings.pkl")
print(data.shape)
print(data.head())
```

```python
X = np.vstack(data["transcript_embedding"].values)
print("Feature matrix shape:", X.shape)
```

```python
y_1m = data["ret_1m"].values
y_3m = data["ret_3m"].values
```

```python
X_train, X_test, y1_train, y1_test = train_test_split(
    X, y_1m, test_size=0.2, random_state=42
)

_, _, y3_train, y3_test = train_test_split(
    X, y_3m, test_size=0.2, random_state=42
)

print(X_train.shape, X_test.shape)
```

```python
# test on linear regression
model_1m = LinearRegression()
model_1m.fit(X_train, y1_train)

pred_1m = model_1m.predict(X_test)
```

```python
r2 = r2_score(y1_test, pred_1m)
corr = np.corrcoef(y1_test, pred_1m)[0,1]

print("R²:", r2)
print("Correlation:", corr)
```

```python
# test on Random forest
rf_1m = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

rf_1m.fit(X_train, y1_train)

rf_pred_1m = rf_1m.predict(X_test)
```

```python
rf_r2 = r2_score(y1_test, rf_pred_1m)
rf_corr = np.corrcoef(y1_test, rf_pred_1m)[0,1]

print("Random Forest R²:", rf_r2)
print("Random Forest Correlation:", rf_corr)
```

```python
# test on XGBoost
xgb_1m = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

xgb_1m.fit(X_train, y1_train)

xgb_pred_1m = xgb_1m.predict(X_test)
```

```python
xgb_r2 = r2_score(y1_test, xgb_pred_1m)
xgb_corr = np.corrcoef(y1_test, xgb_pred_1m)[0,1]

print("XGBoost R²:", xgb_r2)
print("XGBoost Correlation:", xgb_corr)
```
